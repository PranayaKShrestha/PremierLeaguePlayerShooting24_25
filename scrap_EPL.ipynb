```python
from understatapi import UnderstatClient
import pandas as pd
import time
```

```python
client = UnderstatClient()

# Access the EPL league for 2024 season
league = 'EPL'
season = '2024'
players = client.league(league='EPL').get_player_data(season='2024')
df = pd.DataFrame(players)
df[df['id'] == '9272']
```

```python
player_id_list = list(df['id'])
df2 = pd.DataFrame()

for i in player_id_list:
    try:
        df_player = pd.DataFrame(client.player(player=i).get_shot_data())
        df_player = df_player[df_player['season'] == '2024']
        df2 = pd.concat([df2, df_player], ignore_index=True)  # delay to avoid rate limit
    except Exception as e:
        print(f"Error with player {i}: {e}")
        time.sleep(5)  # longer pause before retrying
        
```

```python
epl_2024_25_teams = [
    "Arsenal",
    "Aston Villa",
    "Bournemouth",
    "Brentford",
    "Brighton",
    "Chelsea",
    "Crystal Palace",
    "Everton",
    "Fulham",
    "Ipswich",
    "Leicester",
    "Liverpool",
    "Manchester City",
    "Manchester United",
    "Newcastle United",
    "Nottingham Forest",
    "Southampton",
    "Tottenham",
    "West Ham",
    "Wolverhampton Wanderers"
]

df2_2 = df2[df2['h_team'].isin(epl_2024_25_teams) & df2['a_team'].isin(epl_2024_25_teams)]
```

```python
df2_2_own_goals_removed = df2_2[df2_2['result'] != 'OwnGoal']
```

```python
df2_2_own_goals_removed.to_csv('shot_location_epl_2024.csv')
df.to_csv('players_stats_epl_2024.csv')
```
